```python
import pandas as pd
import ast

df_train = pd.read_csv("training_data.csv")
df_jobs = pd.read_csv("all_job_post.csv")


#  "Python|SQL|ML" -> ["python", "sql", "ml"]
df_train["Skills"] = df_train["Skills"].fillna("")
df_train["skills_clean"] = df_train["Skills"].apply(
    lambda x: [s.strip().lower() for s in x.split("|") if s]
)

df_train_clean = df_train[[
    "skills_clean",
    "Education",
    "Experience Years",
    "Job Role"
]].copy()

df_train_clean.rename(columns={
    "Job Role": "job_role",
    "Education": "education",
    "Experience Years": "experience_years"
}, inplace=True)


df_jobs["job_skill_set"] = df_jobs["job_skill_set"].fillna("[]")

def parse_skills(x):
    try:
        return [s.strip().lower() for s in ast.literal_eval(x)]
    except:
        return []

df_jobs["skills_clean"] = df_jobs["job_skill_set"].apply(parse_skills)

df_jobs_clean = df_jobs[[
    "skills_clean",
    "job_title"
]].copy()

df_jobs_clean.rename(columns={
    "job_title": "job_role"
}, inplace=True)

df_jobs_clean["education"] = None
df_jobs_clean["experience_years"] = None


df_combined = pd.concat([df_train_clean, df_jobs_clean], ignore_index=True)


#empty skills
df_combined = df_combined[df_combined["skills_clean"].map(len) > 0]

df_combined["skills_clean"] = df_combined["skills_clean"].apply(tuple)

#drop duplicates
df_combined = df_combined.drop_duplicates(subset=["skills_clean", "job_role"])


df_combined.to_csv("final_dataset.csv", index=False)

```
